```python
# thanks to this link for doing most of the work

# https://gist.github.com/ImadDabbura/30bf36136221b1ef3ea99586a825c486

# this will run for minutes

# only doing matplotlib for the first one

import cv2
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Read the image
img = cv2.imread(r'YourPath\Dog.jpg',-1)
img_size = img.shape
img = cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
#this last step may not be necessary if your color sceme lines up with the program

# Reshape it to be 2-dimension
X = img.reshape(img_size[0] * img_size[1], img_size[2])

# Run the Kmeans algorithm
km = KMeans(n_clusters=10)
km.fit(X)

# Use the centroids to compress the image
X_compressed = km.cluster_centers_[km.labels_]
X_compressed = np.clip(X_compressed.astype('uint8'), 0, 255)

# Reshape X_recovered to have the same dimension as the original image
X_compressed = X_compressed.reshape(img_size[0], img_size[1], img_size[2])

# Plot the original and the compressed image next to each other
fig, ax = plt.subplots(1, 2, figsize = (12, 8))
ax[0].imshow(img)
ax[0].set_title('Original Image')
ax[1].imshow(X_compressed)
ax[1].set_title('Compressed Image with 10 colors')
for ax in fig.axes:
    ax.axis('off')
plt.tight_layout();

between = cv2.cvtColor(X_compressed,cv2.COLOR_BGR2RGB)

cv2.imwrite(r'YourPath\Dog10.jpg',between)
```

```python
'''
Upper bound in poly-time is O((n^34 log^4(n) k^34 d^8) / σ^6) 
where n is number of points, 
k is number of means, 
d is the diameter of the point set, 
and σ is the standard deviation of Gaussian perturbations

For this example that means that, as the only thing we can control is the number of means --that is k--
as the number of our means go up the computational complexity will go up

for example the difference between computing a k of O(10) vs a k of O(100) increases the complexity of the calculation 34 fold...
'''
```

```python
km1 = KMeans(n_clusters=30)
km1.fit(X)

#will run about 10m

#KMeans(n_clusters = x_1, max_iter = x_2) can save time but result will lose accuracy.

#recommended when you decide to make a mistake and use too many means

X_compressed1 = km1.cluster_centers_[km1.labels_]
X_compressed1 = np.clip(X_compressed1.astype('uint8'), 0, 255)


X_compressed1 = X_compressed1.reshape(img_size[0], img_size[1], img_size[2])

between1 = cv2.cvtColor(X_compressed1,cv2.COLOR_BGR2RGB)

cv2.imwrite(r'YourPath\Dog30.jpg',between1)
```

```python
#will run about 20m

km2 = KMeans(n_clusters=50)
km2.fit(X)


X_compressed2 = km2.cluster_centers_[km2.labels_]

X_compressed2 = np.clip(X_compressed2.astype('uint8'), 0, 255)


X_compressed2 = X_compressed2.reshape(img_size[0], img_size[1], img_size[2])

between2 = cv2.cvtColor(X_compressed2,cv2.COLOR_BGR2RGB)

cv2.imwrite(r'YourPath\Dog50.jpg',between2)
```
